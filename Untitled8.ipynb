```python
# Ejercicio 1.1
print("Ejercicio 1.1:")
total = 30
futbol = 18
baloncesto = 14
ninguno = 5
al_menos_uno = total - ninguno
interseccion = futbol + baloncesto - al_menos_uno

p_interseccion = interseccion / total
p_union = al_menos_uno / total

print(f"a) P(F∩B) = {p_interseccion:.3f}")
print(f"b) P(F∪B) = {p_union:.3f}")
print()
```

```python
# Ejercicio 1.2
print("Ejercicio 1.2:")
# a) Suma 7 o ambos iguales
suma_7 = 6
ambos_iguales = 6
interseccion_1_2 = 0
p_union_1_2 = (suma_7 + ambos_iguales - interseccion_1_2) / 36

# b) Suma par y al menos un 2
suma_par = 18
al_menos_un_2 = 11
interseccion_par_2 = 5
p_interseccion_par_2 = interseccion_par_2 / 36

print(f"a) P(Suma7 ∪ AmbosIguales) = {p_union_1_2:.3f}")
print(f"b) P(SumaPar ∩ AlMenosUn2) = {p_interseccion_par_2:.3f}")
print()
```

```python
print("2. TABLAS CRUZADAS")
print("-" * 40)
```

```python
# Ejercicio 2.1
print("Ejercicio 2.1:")
mujer_fuma = 20
hombre = 120
fuma = 60
hombre_fuma = 40
total_2_1 = 200

p_mujer_fuma = mujer_fuma / total_2_1
p_hombre_union_fuma = (hombre + fuma - hombre_fuma) / total_2_1
p_no_fuma_mujer = 60 / 80  # 60 no fuman de 80 mujeres

print(f"a) P(Mujer ∩ Fuma) = {p_mujer_fuma:.3f}")
print(f"b) P(Hombre ∪ Fuma) = {p_hombre_union_fuma:.3f}")
print(f"c) P(NoFuma|Mujer) = {p_no_fuma_mujer:.3f}")
print()
```

```python
print("3. COMBINACIONES Y PERMUTACIONES")
print("-" * 40)

# Ejercicio 3.1
print("Ejercicio 3.1:")
from math import comb, factorial

total_permutaciones = factorial(5)
permutaciones_juntos = factorial(4) * factorial(2)  # 4! × 2!
p_juntos = permutaciones_juntos / total_permutaciones

print(f"P(A y B juntos) = {p_juntos:.3f}")
```

```python
# Ejercicio 3.2
comites_totales = comb(10, 4)
comites_con_maria_juan = comb(8, 2)  # Elegir 2 de los 8 restantes
p_maria_juan = comites_con_maria_juan / comites_totales

print(f"Ejercicio 3.2: P(María y Juan) = {p_maria_juan:.3f}")
```

```python
# Ejercicio 3.4
print("Ejercicio 3.4:")
total_pares = comb(5, 2)
mismo_color = comb(3, 2) + comb(2, 2)  # 2 rojas + 2 azules
p_mismo_color = mismo_color / total_pares
print(f"P(mismo color) = {p_mismo_color:.3f}")
print()
```

```python
print("4. TEOREMA DE BAYES")
print("-" * 40)
```

```python
# Ejercicio 4.1
print("Ejercicio 4.1:")
p_A = 0.6
p_B = 0.4
p_D_A = 0.02
p_D_B = 0.03

p_D = p_A * p_D_A + p_B * p_D_B
p_A_D = (p_A * p_D_A) / p_D

print(f"P(A|D) = {p_A_D:.3f}")
```

```python
# Ejercicio 4.2
print("Ejercicio 4.2:")
p_E = 0.01
p_T_E = 0.95
p_T_noE = 0.05
p_noE = 0.99

p_T = p_E * p_T_E + p_noE * p_T_noE
p_E_T = (p_E * p_T_E) / p_T

print(f"P(Enfermo|Test+) = {p_E_T:.3f}")
print()
```
