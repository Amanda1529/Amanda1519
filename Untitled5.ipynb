###3.2 Uso de estadísticas descriptivas para identificación de valores extremos
Si bien es cierto el uso de estadísticas será de los resultados que esperamos al final del proceso de limpieza, también son de utilidad para examinar si existen valores extremos.

```python
#Veamos si existen valores extremos en money:
display(df['money'].describe().round(2)) #Redondeamos a dos cifras con round(2)
```

De la tabla anterior se extraen valores como count (total de casos), mean (promedio), std (desviación estándar), min (mínimo), 25%, 50% y 75%  (percentiles) y max (máximo). Su interpretación se deja como ejercicio.

Ahora veamos si existen valores extremos por medio de la fórmula de IQR

```python
#Primero calculamos los percentiles 25 y 75 usando quantile
Q1 = df['money'].quantile(0.25)
Q3 = df['money'].quantile(0.75)
#Segundo calculamos el rango intercuartílico
IQR = Q3 - Q1

#Calculamos los límites de ajuste que son +- 1.5* IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Mostramos los resultados con pritn
print(f'Q1 (Percentil 25): {Q1:.2f}')
print(f'Q3 (Percentil 75): {Q3:.2f}')
print(f'IQR (Rango Intercuartílico): {IQR:.2f}')
print(f'Lower Bound for Outliers (Límite Inferior): {lower_bound:.2f}')
print(f'Upper Bound for Outliers (Límite Superior): {upper_bound:.2f}')

#Ahora, calculamos los outliers o valores extremos mediante un filtro que dice que para la variable money vamo va seleccionar
#los valores por debajo del limite inferior y por encima del límite superior:

outliers = df[(df['money'] < lower_bound) | (df['money'] > upper_bound)]

# Aplicamos un condicional función si si el conjuto está vacío indique no hay
# valore extremos, en el caso contrario que señale cuáles son para tomar una
# decisión.
if outliers.empty:
    print("\n No tenemos valores extremos (outliers) en el conjunto de la variable 'money' basándonos en el método IQR.")
else:
    print("\nEstos son los valore extremos potenciales (outliers) en la variable'money':")
    display(outliers)
```

Q1 (Percentil 25): 27.92
Q3 (Percentil 75): 35.76
IQR (Rango Intercuartílico): 7.84
Lower Bound for Outliers (Límite Inferior): 16.16
Upper Bound for Outliers (Límite Superior): 47.52

 No tenemos valores extremos (outliers) en el conjunto de la variable 'money' basándonos en el método IQR.

##4 Cálculo distribuciones de frecuencias y gráficos
En este apartado calcularemos distribuciones de frecuencias para las variables cualitativas y cuantitativas.

```python
coffee_name_counts = df['coffee_name'].value_counts().reset_index()
coffee_name_counts.columns = ['Tipo de Café', 'Frecuencia']

total_row = pd.DataFrame([['Total', coffee_name_counts['Frecuencia'].sum()]], columns=['Tipo de Café', 'Frecuencia'])
Tabla_Frecuencia= pd.concat([coffee_name_counts, total_row], ignore_index=True)

display(Tabla_Frecuencia)
```

Calculemos la tabla anterior con la columna de porcentajes:

```python
coffee_name_counts = df['coffee_name'].value_counts().reset_index()
coffee_name_counts.columns = ['Tipo de Café', 'Frecuencia']

total_frequency = coffee_name_counts['Frecuencia'].sum()
coffee_name_counts['Porcentaje'] = (coffee_name_counts['Frecuencia'] / total_frequency) * 100

total_row = pd.DataFrame([['Total', total_frequency, 100.00]], columns=['Tipo de Café', 'Frecuencia', 'Porcentaje'])
Tabla_Frecuencia = pd.concat([coffee_name_counts, total_row], ignore_index=True)

display(Tabla_Frecuencia.round(2))
```

Con base en estos datos elaboraremos un gráfico de barras horizontales

```python
# Excluimos el total en la tabla anterior porque no nos interesa para la gráfica
Tabla_Frecuencia_plot = Tabla_Frecuencia[Tabla_Frecuencia['Tipo de Café'] != 'Total'].copy()

# Ordenamos la tabla de frecuencia en orden descendente
Tabla_Frecuencia_plot = Tabla_Frecuencia_plot.sort_values('Frecuencia', ascending=False)
#Agregamos los elementos del gráfico:
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Porcentaje', y='Tipo de Café', data=Tabla_Frecuencia_plot, color='skyblue') # Colocamos el color cielo "skyblue"
plt.title('Distribución de frecuencia del tipo de café solicitado por los clientes') #Título
plt.xlabel('Porcentaje') #El porcentaje
plt.ylabel('Tipo de Café') #Las categorías
plt.tight_layout()

# Agregamos las etiquetas
for index, row in Tabla_Frecuencia_plot.iterrows():
    ax.text(row['Porcentaje'], index, f'{row["Porcentaje"]:.2f}%', va='center')

plt.show()
```

Haremos algo similar para "Time of day"

```python
horarios = df['Time_of_Day'].value_counts().reset_index()
horarios.columns = ['Horario del día', 'Frecuencia']

total_frequency = horarios['Frecuencia'].sum()
horarios['Porcentaje'] = (horarios['Frecuencia'] / total_frequency) * 100

total_row = pd.DataFrame([['Total', total_frequency, 100.00]], columns=['Horario del día', 'Frecuencia', 'Porcentaje'])
Tabla_Frecuencia_horario = pd.concat([horarios, total_row], ignore_index=True)

display(Tabla_Frecuencia_horario.round(2))
```

```python
# Excluimos el total de la tabla
Tabla_Frecuencia_horario_plot = Tabla_Frecuencia_horario[Tabla_Frecuencia_horario['Horario del día'] != 'Total'].copy()

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Horario del día', y='Porcentaje', data=Tabla_Frecuencia_horario_plot, color='skyblue')
plt.title('Distribución de frecuencia por horario del día en porcentaje')
plt.xlabel('Horario del día')
plt.ylabel('Porcentaje')
plt.tight_layout()

# Agregamos las etiquetas
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

Graficaremos a "Weekdaysort"	y "Monthsort" que son los días de la semana y meses del año ya ordenados

```python
# Calculamos la distribución de frecuencias para Weekdaysort
weekday_counts = df['Weekdaysort'].value_counts().reset_index()
weekday_counts.columns = ['Weekday Sort', 'Frecuencia']

# Ordenamos por los días
weekday_counts = weekday_counts.sort_values('Weekday Sort')

# Calculamos los porcentajes
total_frequency = weekday_counts['Frecuencia'].sum()
weekday_counts['Porcentaje'] = (weekday_counts['Frecuencia'] / total_frequency) * 100

# Creamos el gráfico de barras verticales variable temporal
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Weekday Sort', y='Porcentaje', data=weekday_counts, color='skyblue') # Use percentages on y-axis
plt.title('Distribución de frecuencia por día de la semana en porcentaje') # Update title
plt.xlabel('Día de la semana (ordenado)')
plt.ylabel('Porcentaje') # Update y-axis label
plt.tight_layout()

# Agregamos las etiquetas
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()

display(weekday_counts.round(2))
```

```python
# Calculate frequency distribution for Monthsort
month_counts = df['Monthsort'].value_counts().reset_index()
month_counts.columns = ['Month Sort', 'Frecuencia']

# Sort by Month Sort to ensure correct order of months
month_counts = month_counts.sort_values('Month Sort')

# Calculate percentages
total_frequency = month_counts['Frecuencia'].sum()
month_counts['Porcentaje'] = (month_counts['Frecuencia'] / total_frequency) * 100

# Create line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month Sort', y='Porcentaje', data=month_counts, marker='o') # Use lineplot and add markers
plt.title('Distribución de frecuencia por mes en porcentaje') # Update title
plt.xlabel('Mes (ordenado)')
plt.ylabel('Porcentaje') # Update y-axis label
plt.grid(True) # Add a grid for better readability
plt.show()

display(month_counts.round(2))
```

Como se puede visualizar los meses de año  con mayor porcentaje de compras de café por parte de los clientes son en marzo, febrero y octubre respectivamente.

```python
# Calculate frequency distribution for Monthsort
month_counts = df['Monthsort'].value_counts().reset_index()
month_counts.columns = ['Month Sort', 'Frecuencia']

# Sort by Month Sort to ensure correct order of months
month_counts = month_counts.sort_values('Month Sort')

# Calculate percentages
total_frequency = month_counts['Frecuencia'].sum()
month_counts['Porcentaje'] = (month_counts['Frecuencia'] / total_frequency) * 100

# Map numerical months to Spanish month names
month_names_spanish = {
    1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril',
    5: 'Mayo', 6: 'Junio', 7: 'Julio', 8: 'Agosto',
    9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'
}
month_counts['Month Name'] = month_counts['Month Sort'].map(month_names_spanish)

# Create line chart
plt.figure(figsize=(12, 6)) # Increase figure size for better readability of month names
sns.lineplot(x='Month Name', y='Porcentaje', data=month_counts, marker='o') # Use month names on x-axis
plt.title('Distribución de frecuencia por mes en porcentaje')
plt.xlabel('Mes') # Update x-axis label
plt.ylabel('Porcentaje')
plt.grid(True)
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

display(month_counts.round(2))
```

Ahora para la variable Money que es continua

```python
# Define bins for the money variable, excluding ranges with 0 frequency
bins = [0, 20, 25, 30, 35, 40] # Adjusted bins

# Create a new column with the bin each transaction falls into
df['money_bin'] = pd.cut(df['money'], bins=bins, right=False)

# Calculate frequency distribution by bin
money_bin_counts = df['money_bin'].value_counts().reset_index()
money_bin_counts.columns = ['Money Range', 'Frecuencia']

# Sort by Money Range (optional, but good for readability)
money_bin_counts = money_bin_counts.sort_values('Money Range')

# Calculate percentages
total_frequency = money_bin_counts['Frecuencia'].sum()
money_bin_counts['Porcentaje'] = (money_bin_counts['Frecuencia'] / total_frequency) * 100

# Display the grouped frequency distribution table
display(money_bin_counts.round(2))
```

Para efectos del ejercicio haremos un caso y mostraremos un gráfico de calor de forma ilustrativa.
- Usualmente se debe tener claridad del tipo de variable que se usará para calcular la correlación.
- Debe existir teoría que sustente este tipo de gráficos.
- Existen varios tipos de correlación entre variables esto depende del tipo de variables que se estén considerando.

```python
# Seleccionamos las variables numéricas
numerical_df = df.select_dtypes(include=np.number)

# Calculamos la matriz de correlaciones ** Esto solo con un fin de ejemplo ya que se debe tener evidencia de que las correlaciones existan y algunas variables no son continuas***
correlation_matrix = numerical_df.corr()

# Creamos un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de calor de la correlación de las variables numéricas en el archivo de datos')
plt.show()
```
