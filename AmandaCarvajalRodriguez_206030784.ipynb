#Pregunta 1

```python
# Datos
total = 60
futbol = 30
baloncesto = 25
ambos = 10  # Intersección F y B

# Cálculo de la unión
union = futbol + baloncesto - ambos

# Probabilidad de la unión
prob = union / total

# Mostrar resultado con 4 decimales
print(round(prob, 4))

# Respuesta 5: Exclusivamente fútbol
solo_futbol = futbol - ambos
print(f"Respuesta 5 (Exclusivamente fútbol): {solo_futbol}")
```

#Pregunta 2

```python
from scipy.stats import binom

n, p = 50, 0.1

# Determinación de la probabilidad

prob = binom.pmf(5, n, p) # Example: probability of exactly 5 successes
print(round(prob, 4))

# P(X > 40) es lo mismo que 1 - P(X <= 40)
prob_mas_de_40 = 1 - binom.cdf(40, n, p)
print(f"Probabilidad de más de 40: {prob_mas_de_40}")

# P(X = 4)
prob_exacto_4 = binom.pmf(4, n, p)
print(f"Probabilidad de exactamente 4: {round(prob_exacto_4, 4)}")
```

#Pregunta 3


```python
from scipy.stats import norm

##Datos X: Estaturas de estudiantes, X ~ N(170,10)

media, desv = 170, 10
x = 185 # Definir x como 185 para las siguientes operaciones

# Integración de la superficie bajo la curva correspondiente a la cola superior derecha

prob_mas185 = norm.sf(185, loc=media, scale=desv)
print(round(prob_mas185, 4))

# Respuesta 3: Valor Z correspondiente
z_score = (x - media) / desv
print(f"Respuesta 3 (Valor de Z): {round(z_score, 1)}")

# Respuesta 5: Percentil de 185 cm
# El percentil es la probabilidad acumulada (cdf) multiplicada por 100
percentil = norm.cdf(x, loc=media, scale=desv) * 100
print(f"Respuesta 5 (Percentil): {int(round(percentil, 0))}")
```

#Pregunta 4


```python
#Datos
#E: estar enfermo, noE: no estar enfermo, pos: positivo, neg: negativo

P_E = 0.01      #Prevalencia
P_pos_E = 0.95   #Sensibilidad
P_neg_noE = 0.90 #Especificidad

##Pasos

# 1. Calculamos la probabilidad de que las personas sanas reciben un valor positivo por error del test (falso positivo)
# P(+ | no E)

P_pos_noE = 1 - P_neg_noE

##2. Calculamos las probabilidades del numerador y denominador:

numerador = P_E * P_pos_E #Probabilidad de estar enfermo y dar positivo
denominador = numerador + (1 - P_E) * P_pos_noE ##Probabilidad total

# Aplicación del Teorema de Bayes para la probabilidad posterior

prob_bayes = numerador / denominador
print(round(prob_bayes, 4))

# Respuesta 4: Probabilidad de que un sano dé positivo
print(f"Probabilidad de que un sano dé positivo: {round(P_pos_noE, 1)}")
```
